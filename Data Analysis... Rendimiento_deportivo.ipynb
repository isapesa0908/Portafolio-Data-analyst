## Análisis de Datos de Rendimiento en la NBA: Tendencias y Patrones Competitivos


Un estudio analítico sobre el impacto de localía, rendimiento ofensivo y comportamiento histórico de los equipos en la NBA.

### PREGUNTAS OBJETIVO:

1. ¿ Como se han ganado mas partidos de local o visitante?
2. ¿Que equipos han hecho la mayor cantidad de puntos, tanto en casa como de visitantente ?
3. ¿Cuantos partidos han ganado los distintos equipos estando en casa o de visitante?
4. ¿Como se han comportado los puntos anotados a lo largo de los años?
5. ¿Contra caules equipos han ganado mas estado en casa y visitante?

```python
#Importamos las librerias que utilizaremos 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
games = pd.read_csv('C:/Users/isape/OneDrive/Documentos/Panda/games.csv')
games.info()
```

```python
games = games[['GAME_DATE_EST', 'GAME_ID', 'GAME_STATUS_TEXT', 'TEAM_ID_home',
               'TEAM_ID_away', 'PTS_home', 'PTS_away', 'HOME_TEAM_WINS']]
games.head()
```

```python
games['GAME_DATE_EST'] = pd.to_datetime(games['GAME_DATE_EST'])
```

```python
games.info()
```

```python
games['year'] = games['GAME_DATE_EST'].dt.year
games.head()
```

```python
# Cargamos los datos Teams:
teams = pd.read_csv('C:/Users/isape/OneDrive/Documentos/Panda/teams.csv')
teams.head()
```

```python

teams = teams[['TEAM_ID', 'CITY', 'NICKNAME']]
teams.head()
```

```python
#Creamos un nuevo conjunto de datos, con la union de los anteriores conjuntos:
home_games = pd.merge(games, teams, left_on='TEAM_ID_home', right_on='TEAM_ID', how='inner')
home_games.head()
```

```python
# Renombramos columnas:
home_games = home_games.rename(columns={
    'CITY':'city_home',
    'NICKNAME': 'nickname_home'    
})
home_games.head()
```

```python
#Repetimos el proceso con la infomacion de visitante:
both_games = pd.merge(home_games, teams, left_on='TEAM_ID_away', right_on='TEAM_ID', how='inner')
both_games.head()
```

```python
# Renombramos las nuevas columnas:
both_games = both_games.rename(columns={
    'CITY':'city_away',
    'NICKNAME': 'nickname_away'    
})
both_games.head()
```

```python
#Eliminamos las columnas:
both_games = both_games.drop(columns=[
    'TEAM_ID_home', 'TEAM_ID_away', 'TEAM_ID_x', 'TEAM_ID_y'
])
both_games.head()
```

```python
#Creamos una copia del conjunto de datos lo renombramos y hacemos que las letras pasen a minusculas:
games = both_games.copy()
games.columns = games.columns.str.lower()
games.head()
```

```python
games['total_points'] = games['pts_home'] + games['pts_away']
games.head()
```

```python
#Los 5 partidos con mas puntos anotados
games.nlargest(n=5, columns = 'total_points')
```

```python
#Los 5 partidos con menos puntos anotados
games.nsmallest(n=5, columns = 'total_points')
```

# Graficos:

```python
total = games['home_team_wins'].value_counts()
total
```

```python
plt.pie(total.values, labels=['Casa', 'Visitante'], autopct= '%.0f%%')
plt.show()
```

```python
#Cantidad de puntos por equipo cuando juegan de local o visitante:
plt.figure(figsize=(20, 6))
home_wins = games[games['home_team_wins'] == 1 ] 
result = home_wins.groupby('nickname_home')['pts_home'].aggregate(np.mean).reset_index().sort_values('pts_home')

ax = sns.barplot(data=home_wins, x = 'nickname_home', y = 'pts_home', order = result ['nickname_home'], errorbar=None)
ax.bar_label(ax.containers[-1], fmt='%.0f', label_type='center', rotation = 45) 
plt.xticks(rotation = 75)
plt.show()
```

```python
plt.figure(figsize=(20, 6))
away_wins = games[games['home_team_wins'] == 0 ] 
result = home_wins.groupby('nickname_away')['pts_away'].aggregate(np.mean).reset_index().sort_values('pts_away')

ax = sns.barplot(data =away_wins, x = 'nickname_away', y = 'pts_away', order = result ['nickname_away'], errorbar=None)
ax.bar_label(ax.containers[-1], fmt='%.0f', label_type='center', rotation = 45) 
plt.xticks(rotation = 75)
plt.show()
```
